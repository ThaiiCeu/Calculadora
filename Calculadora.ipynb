#Calculadora!

```python
while True:

#Opções da calculadora:
    print("\nCalculadora!")
    print("\nSelecione a operação:")
    print("1. Adição")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")
    print("5. Sair")

#Colocando variáveis:
    choice = input("Digite sua escolha (1/2/3/4/5): ")
    num1 = input("Adicione um número:")
    num1 = float(num1)

    num2 = input("Adicione um número:")
    num2 = float(num2)

#Para finalizar a calculadora
    if choice == '5':
      print("Encerrando a calculadora. Até mais!")
      break

#Soma
    elif choice == '1':
      result = num1 + num2
      print(f"{num1} + {num2} = {result}")

#Subtração
    elif choice == '2':
      result = num1 - num2
      print(f"{num1} - {num2} = {result}")

#Multiplicação
    elif choice == '3':
      result = num1 * num2
      print(f"{num1} * {num2} = {result}")

#Divisão
    elif choice == '4':
      if num2 == 0:
        print("Erro! Denominador é zero.")

    else:
          result = num1 / num2
          print(f"{num1} / {num2} = {result}")

    break #Para parar o código
```
